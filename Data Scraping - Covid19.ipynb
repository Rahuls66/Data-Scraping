```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Covid Cases as on 11th April 2021

url = "https://api.covid19india.org/data.json"
```

```python
html = requests.get(url).json()
```

```python
html
```

```python
html.keys()
```

```python
html['statewise']
```

```python
#Using iter() function to iterate over list of values of key 'statewise'
#Using next(it) to skip first iteraion (of Total Covid Cases of India)

active = []
confirmed = []
deaths = []
recovered = []
state = []

it = iter(html['statewise'])
next(it)

for i in it:
    state.append(i['state'])
    active.append(i['active'])
    confirmed.append(i['confirmed'])
    deaths.append(i['deaths'])
    recovered.append(i['recovered'])
pd.DataFrame({'State':state, 'Active':active, 'Confirmed':confirmed, 'Deaths':deaths, 'Recovered':recovered})
```

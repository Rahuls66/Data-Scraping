```python
browser = webdriver.Chrome()
```

```python
#Function to get Lists of All Titles, Prices, Descriptions and Ratings For given URL
```

```python
def laptop(soup):
    title = [i.text.strip() for i in soup.find_all('a', class_='title')]
    price = [i.text.strip() for i in soup.find_all('h4', class_='pull-right price')]
    descr = [i.text.strip() for i in soup.find_all('p', class_='description')]
    ratings = [len(i('span')) for i in soup.find_all('div', class_='ratings')]
    return pd.DataFrame({'Title':title, 'Price':price, 'Description':descr, 'Ratings':ratings})
```

```python
#For 10 Pages
```

```python
stub = 'https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page={}'
```

```python
from tqdm import tqdm
import time

data = []
for i in tqdm(range(1,11)):
    url = stub.format(i)
    browser.get(url)
    time.sleep(2)
    html = browser.page_source    
    soup = BeautifulSoup(html, 'html.parser')
    data.append(laptop(soup))
```

```python
df_laptop = pd.concat(data, axis = 0)
```

```python
df_laptop = df_laptop.reset_index(drop = True)
```

```python
def job(soup):
    jobRole = [i.text for i in soup.find_all('a', class_='title fw500 ellipsis')]
    comp = [i.text for i in soup.find_all('a', class_='subTitle ellipsis fleft')]
    exp = [i('li')[0].text for i in soup.find_all('ul', class_='mt-7')]
    sal = [i('li')[1].text for i in soup.find_all('ul', class_='mt-7')]
    loc = [i('li')[2].text for i in soup.find_all('ul', class_='mt-7')]
    return pd.DataFrame({'Job':jobRole, 'Company':comp, 'Experience':exp, 'Salary':sal, 'Location':loc})
```

```python
job(soup)
```

```python
stub = "https://www.naukri.com/data-science-jobs-in-bangalore-bengaluru-{}?k=data%20science&l=bangalore%2Fbengaluru"
```

```python
from tqdm import tqdm
import time

data = []
for i in tqdm(range(1,11)):
    url = stub.format(i)
    browser.get(url)
    time.sleep(5)
    html = browser.page_source    
    soup = BeautifulSoup(html, 'html.parser')
    data.append(job(soup))
```

```python
df = pd.concat(data, axis = 0)
```

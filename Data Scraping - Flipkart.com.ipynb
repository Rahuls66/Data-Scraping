```python
def review(soup):
    revTitle = [i.text for i in soup.find_all('p', class_='_2-N8zT')]
    rev = [i.text for i in soup.find_all('div', class_='t-ZTKy')]
    stars = [i.text for i in soup.find_all("div", class_="_3LWZlK _1BLPMq")]
    
    return pd.DataFrame({'Title':revTitle, 'Review':rev, 'Stars':stars})
```

```python
review(soup)
```

```python
stub = "https://www.flipkart.com/apple-iphone-12-blue-128-gb/product-reviews/itm02853ae92e90a?pid=MOBFWBYZKPTZF9VG&lid=LSTMOBFWBYZKPTZF9VG6GMIFT&marketplace=FLIPKART&page={}"
```

```python
from tqdm import tqdm
import time

data = []
for i in tqdm(range(1, 6)):
    url = stub.format(i)
    browser.get(url)
    time.sleep(5)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    data.append(review(soup))
```

```python
df = pd.concat(data, axis  = 0)
```
